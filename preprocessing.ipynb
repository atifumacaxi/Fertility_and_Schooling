```python
import pandas as pd
import numpy as np
```

```python
df_fertility = pd.read_csv('./fertility_rate.csv')
df_fertility.head()
```

```python
df_schooling = pd.read_csv('mean-years-of-schooling-long-run.csv', sep=';')
df_schooling.head()
```

```python
df_schooling = df_schooling.rename(columns={'Entity': 'Country'})
df_schooling
```

```python
df_schooling = df_schooling.drop(columns='Code')
df_schooling.head()
```

```python
df_fertility = pd.melt(df_fertility,
                       id_vars=["Country",],
                       var_name="Year", value_name="fertility")
df_fertility.head()
```

```python
df_fertility = df_fertility.sort_values(["Country"])
df_fertility
```

```python
df_schooling_fertility = pd.concat([df_schooling, df_fertility])
df_schooling_fertility
```
